3. Explain the following termsg
M The difference in np.array, np.asarray and np.asanyarrayX
M The difference between Deep copy and shallow copy

Difference Between np.array, np.asarray, and np.asanyarray
np.array()

Always creates a new NumPy array from any input (like lists or tuples).

Even if the input is already an array, it will still create a new one unless copy=False.

np.asarray()

Converts the input into a NumPy array only if itâ€™s not already one.

More efficient when working with existing arrays, since it avoids unnecessary copies.

np.asanyarray()

Similar to np.asarray() but it also preserves NumPy subclasses, such as np.matrix or masked arrays.

Useful when you want to keep special behaviors associated with subclasses.

Key Point:
Use asarray for efficiency and asanyarray when working with custom array types.

Difference Between Deep Copy and Shallow Copy
Shallow Copy

Only copies the reference to the original object, not the data itself.

If you change the original array, the shallow copy reflects those changes.

Deep Copy

Makes a completely independent copy of the object and its data.

Changes to one do not affect the other.

Key Point:
Use .copy() in NumPy or copy.deepcopy() in general Python when you want a deep, independent copy of the data.

5. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from arrayX

```python
import numpy as np

# Create a (5, 6) array of random integers between 1 and 10 (inclusive)
array = np.random.randint(1, 11, size=(5, 6))

# Extract even integers
even_integers = array[array % 2 == 0]

# Extract odd integers
odd_integers = array[array % 2 != 0]

# Display the results
print("Original Array:\n", array)
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

6.  Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the
following operations:

7. Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and
convert it to a numeric data type. Also display the table attributes and data types of each column.

```python
import pandas as pd

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Phone': ['+44-7911-123456', '(020) 7946 0958', '07911 234567'],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com']
}

# Create DataFrame
df = pd.DataFrame(data)

# Clean 'Phone' column to keep only numeric characters
df['Phone'] = df['Phone'].str.replace(r'\D', '', regex=True)

# Convert to numeric type (optional: you can also keep it as string if leading zeros are important)
df['Phone'] = pd.to_numeric(df['Phone'])

# Display cleaned DataFrame
print("Cleaned DataFrame:\n", df)

# Display DataFrame info (attributes and data types)
print("\nData Types and Attributes:")
print(df.info())
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate two random arrays of 100 points each
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame with the x and y arrays
data = pd.DataFrame({'x': x, 'y': y})

# Create a new column 'quadrant' based on the position of each point relative to the origin
data['quadrant'] = np.where((data['x'] >= 0) & (data['y'] >= 0), 'Quadrant 1',
                            np.where((data['x'] < 0) & (data['y'] >= 0), 'Quadrant 2',
                                     np.where((data['x'] < 0) & (data['y'] < 0), 'Quadrant 3', 'Quadrant 4')))

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', data=data, hue='quadrant', palette='Set1', s=100)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)  # 100 points from 0 to 4*pi
y = np.sin(x)  # Sine of each x value

# Create a Bokeh plot
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')

# Plot the sine wave as a line
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Add grid lines
p.grid.visible = True

# Show the plot
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=5)  # Random values between 1 and 20

# Create a DataFrame for better handling of the data
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(data)

# Create a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Create the bar chart
p.vbar(x='Category', top='Value', width=0.9, source=source, legend_field="Category",
       line_color="white", color="skyblue", fill_color="skyblue")

# Add hover tool to display the exact value
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Set axis labels
p.xaxis.axis_label = "Category"
p.yaxis.axis_label = "Value"

# Set properties for the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
output_file("bar_chart.html")
show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'8

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)  # 100 equally spaced points between 0 and 10
y = np.random.rand(100)  # Random y-values between 0 and 1

# Create the figure with a line plot
fig = go.Figure()

# Add a line plot to the figure
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update the layout with title and axis labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    template='plotly_dark'  # Optional: To set the background to dark
)

# Show the plot
fig.show()
```
