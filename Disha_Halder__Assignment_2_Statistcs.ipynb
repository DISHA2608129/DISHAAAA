1. What is hypothesis testing in statistics?

Hypothesis testing in statistics is a method used to determine whether there is enough statistical evidence to support a particular hypothesis about a population based on sample data. It involves making an assumption (hypothesis) about a population parameter, collecting data, and then using statistical techniques to test whether the data supports or contradicts the hypothesis.

Here’s how the process typically works:

Formulate the Hypotheses:

Null Hypothesis (H₀): This is the default assumption, usually stating that there is no effect or no difference. It is what you aim to test against.

Alternative Hypothesis (H₁ or Ha): This is the hypothesis that contradicts the null hypothesis, stating that there is an effect or a difference.

Select the Significance Level (α): This is the probability of rejecting the null hypothesis when it is actually true. It is typically set at 0.05 or 5%, meaning there's a 5% chance of committing a Type I error (false positive).

Collect and Analyze Data: Data is collected from the sample, and statistical methods are applied to compute test statistics (e.g., t-test, chi-square test, z-test) depending on the nature of the data.

Calculate the P-value: The p-value represents the probability of obtaining test results at least as extreme as the results actually observed, under the assumption that the null hypothesis is true. If the p-value is less than the significance level (α), the null hypothesis is rejected.

Make a Decision:

If p-value ≤ α, reject the null hypothesis (support for the alternative hypothesis).

If p-value > α, fail to reject the null hypothesis (no sufficient evidence to support the alternative hypothesis).

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H₀) is a statement or assumption about a population parameter that suggests there is no effect, no difference, or no relationship between variables. It represents the status quo or a default position that researchers aim to test against.

The alternative hypothesis (H₁ or Ha), on the other hand, is the hypothesis that contradicts the null hypothesis. It proposes that there is an effect, a difference, or a relationship between variables. Researchers use the alternative hypothesis to explore potential findings that are contrary to the null hypothesis.

Key Differences:
Purpose:

Null Hypothesis (H₀): It typically suggests no effect, no difference, or no relationship. For example, "The mean height of women is 65 inches."

Alternative Hypothesis (H₁): It suggests that there is a meaningful effect or difference. For example, "The mean height of women is not 65 inches."

Testing Focus:

The null hypothesis is tested to see if there is enough evidence to reject it in favor of the alternative hypothesis.

The alternative hypothesis is what researchers hope to prove with evidence collected from the sample.

Statistical Decision:

If the data provides strong enough evidence against the null hypothesis, it is rejected in favor of the alternative hypothesis.

If the data does not provide sufficient evidence, the null hypothesis is not rejected (it is not necessarily "proven" true).

3. What is the significance level in hypothesis testing, and why is it important?

The significance level (denoted as α) in hypothesis testing is a threshold that helps determine whether the evidence from the sample data is strong enough to reject the null hypothesis. It is the probability of making a Type I error, which occurs when the null hypothesis is incorrectly rejected (i.e., a false positive).

Importance of the Significance Level:
Error Control: The significance level controls the likelihood of committing a Type I error—rejecting the null hypothesis when it is actually true. By setting a significance level (often 0.05), researchers limit the probability of making this error to 5%. In other words, there's a 5% chance of concluding there's an effect or difference when, in fact, there is none.

Decision Threshold: The significance level determines the p-value threshold below which the null hypothesis is rejected. If the p-value (the probability of obtaining a test statistic as extreme as the observed one, assuming the null hypothesis is true) is smaller than α, the null hypothesis is rejected.

4. What does a P-value represent in hypothesis testing?

The p-value helps you measure the strength of evidence against the null hypothesis (H₀):

A small p-value (typically ≤ 0.05) indicates strong evidence against H₀, so you reject the null hypothesis.

A large p-value (typically > 0.05) suggests weak evidence against H₀, so you fail to reject the null hypothesis.


Suppose it is testing whether a new medication lowers blood pressure more than the current standard.

H₁ (alternative hypothesis): The new medication lowers blood pressure more than the current one.

After running a test, you get a p-value of 0.02.

Since 0.02 is less than 0.05, you would reject H₀ and conclude there's significant evidence the new medication is more effective.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Type I and Type II Errors in Hypothesis Testing
In hypothesis testing, there are two types of errors that can occur when making decisions about the null hypothesis (H₀). These are known as Type I and Type II errors.

Type I Error (False Positive)
A Type I error occurs when you reject the null hypothesis (H₀) when it is actually true. This means that you conclude there is an effect or difference when, in reality, there isn't. For example, if you're testing a new drug and conclude it works better than the current drug (rejecting H₀), but the new drug actually has no significant effect, you've made a Type I error. The risk of making a Type I error is represented by the significance level (α), which is typically set to 0.05. This means there is a 5% chance of rejecting the null hypothesis when it is true.

Type II Error (False Negative)
A Type II error happens when you fail to reject the null hypothesis (H₀) when it is actually false. In this case, you miss detecting a real effect or difference. For example, if the new drug actually works but you fail to reject the null hypothesis and conclude it doesn't work better than the current drug, you've made a Type II error. The probability of making a Type II error is represented by β, and the power of a test is calculated as 1 - β. A high power increases the likelihood of detecting a true effect.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One-Tailed Test
Definition: A one-tailed test is used when the alternative hypothesis (H₁) predicts a specific direction of the effect. That is, you're testing if a parameter is either greater than or less than a certain value, but not both.

Example: Suppose you're testing a new drug and hypothesize that it will improve patient outcomes. In this case, you might use a one-tailed test to test whether the drug is better than the existing treatment (i.e., you’re only interested in whether the drug leads to improvement, not if it’s worse). The alternative hypothesis (H₁) would be that the new drug leads to better outcomes (greater than the current treatment).

Critical Region: The critical region (where you reject the null hypothesis) is only on one side of the distribution (either the left tail or right tail), depending on the direction of the hypothesis.

Example Hypotheses:

Null Hypothesis (H₀): The new drug has no effect (or is worse).

Alternative Hypothesis (H₁): The new drug is better (greater effect).

Two-Tailed Test
Definition: A two-tailed test is used when the alternative hypothesis (H₁) predicts that the parameter is different from the hypothesized value, but does not specify the direction of the difference. It tests for the possibility of an effect in either direction (greater than or less than).

Example: If you're testing the same new drug, but you don’t know whether it will improve or worsen patient outcomes, you’d use a two-tailed test. Here, you are testing whether the new drug differs from the current treatment in either direction (better or worse). The alternative hypothesis (H₁) would be that the new drug causes a difference in patient outcomes (either greater or lesser).

Critical Region: The critical regions (where you reject the null hypothesis) are on both sides of the distribution. This means you reject H₀ if the test statistic falls into either the left or right tail.

19. What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is a continuous probability distribution that arises in statistical tests involving the ratio of variances. It is particularly used in analysis of variance (ANOVA), regression analysis, and testing the equality of variances between two or more groups.

Key Characteristics of the F-distribution:
Shape: The F-distribution is positively skewed (i.e., it has a long right tail), and it is always non-negative, as it represents a ratio of variances.

Degrees of Freedom: The shape of the F-distribution is determined by two degrees of freedom—one for the numerator and one for the denominator. The numerator degrees of freedom correspond to the variance of the sample(s) being tested, and the denominator degrees of freedom correspond to the error variance or the pooled variance of the data.

Non-Symmetry: Unlike the normal distribution, the F-distribution is not symmetrical. The shape of the distribution depends on the specific degrees of freedom of both the numerator and denominator.

When is the F-distribution used in hypothesis testing?
The F-distribution is used in hypothesis testing when you are dealing with the ratio of variances between two or more groups. Here are common situations in which the F-distribution is used:

Analysis of Variance (ANOVA):

ANOVA is used to compare the means of three or more groups to see if there is a significant difference between them. The F-distribution helps determine whether the observed differences in means are greater than what could be expected by random variation alone.

The null hypothesis in ANOVA states that all group means are equal, while the alternative hypothesis states that at least one group mean is different.

Testing Equality of Variances:

The F-test is used to test whether two population variances are equal. In this case, the null hypothesis is that the variances of two populations are equal, and the alternative hypothesis is that they are not equal.

The F-statistic is calculated as the ratio of the two sample variances, and it follows an F-distribution.

Regression Analysis:

In regression analysis, the F-test is used to assess the overall significance of the model. It tests whether at least one of the predictors in a multiple regression model has a non-zero coefficient (i.e., the model explains a significant portion of the variance in the dependent variable).

The null hypothesis is that all regression coefficients are equal to zero, meaning that the model does not explain any variability in the dependent variable. The alternative hypothesis is that at least one coefficient is not equal to zero.

21. What are the different types of ANOVA tests?

There are several types of ANOVA (Analysis of Variance) tests, each used for different research scenarios based on the number of independent variables and the specific research question. Here are the main types of ANOVA tests:

1. One-Way ANOVA
Purpose: Used when there is one independent variable (factor) with more than two levels (groups), and you want to test whether there is a significant difference between the means of the groups.

Example: Comparing the effectiveness of three different diets on weight loss. The independent variable is the type of diet (3 levels), and the dependent variable is the weight loss (measurements).

Hypotheses:

Null Hypothesis (H₀): The means of all groups are equal.

Alternative Hypothesis (H₁): At least one group mean is different from the others.

Assumptions: Independence, normality, and homogeneity of variance (equal variances across groups).

2. Two-Way ANOVA
Purpose: Used when there are two independent variables (factors) to examine their individual effects on the dependent variable as well as any interaction effects between the two variables.

Example: Investigating the effects of both diet type (3 levels) and exercise intensity (2 levels) on weight loss (dependent variable).

Hypotheses:

Null Hypothesis (H₀): No main effect of diet, no main effect of exercise intensity, and no interaction effect.

Alternative Hypothesis (H₁): At least one of the main effects or the interaction effect is significant.

Assumptions: Independence, normality, and homogeneity of variance for each combination of the two independent variables.

3. Three-Way ANOVA
Purpose: Used when there are three independent variables, allowing for the examination of their individual effects as well as their interactions with each other.

Example: Investigating the effects of diet type (3 levels), exercise intensity (2 levels), and age group (2 levels) on weight loss.

Hypotheses:

Null Hypothesis (H₀): No main effects for each factor, no two-way interactions, and no three-way interaction.

Alternative Hypothesis (H₁): At least one main effect or interaction effect is significant.

Assumptions: Independence, normality, and homogeneity of variance for each combination of the three independent variables.

4. Repeated Measures ANOVA
Purpose: Used when the same subjects are used for multiple measurements (i.e., repeated observations), allowing the analysis of how the dependent variable changes over time or under different conditions.

Example: Measuring the effect of three different types of medication on blood pressure at three time points: before treatment, after 1 week, and after 2 weeks.

Hypotheses:

Null Hypothesis (H₀): The means for all time points or conditions are equal.

Alternative Hypothesis (H₁): At least one time point or condition has a significantly different mean.

Assumptions: Normality, sphericity (equal variances of the differences between conditions), and the independence of observations within each condition.

5. Multivariate Analysis of Variance (MANOVA)
Purpose: Extends the One-Way or Two-Way ANOVA to multiple dependent variables. It is used to assess whether changes in the independent variables significantly affect a combination of dependent variables.

Example: Investigating the effects of different teaching methods (independent variable) on students' scores in math and science (dependent variables).

Hypotheses:

Null Hypothesis (H₀): The means of the dependent variables are equal across all groups.

Alternative Hypothesis (H₁): At least one of the dependent variables differs across groups.

Assumptions: Multivariate normality, homogeneity of covariance matrices, and independence of observations.

6. Welch’s ANOVA
Purpose: An alternative to the standard One-Way ANOVA when the assumption of homogeneity of variances (equal variances across groups) is violated. It is used when the variances of the groups are unequal.

Example: Testing whether three different teaching methods have different effects on student performance, where the variances between the groups are unequal.

Assumptions: Independence of observations, normality, and unequal variances (homogeneity of variances is not assumed).

7. Kruskal-Wallis H Test
Purpose: A non-parametric alternative to the One-Way ANOVA, used when the assumptions of normality and homogeneity of variances are not met. It is used to compare the medians across more than two groups.

Example: Comparing the effect of three different diets on weight loss when the data is not normally distributed.

Hypotheses:

Null Hypothesis (H₀): The distributions of all groups are equal.

Alternative Hypothesis (H₁): At least one group differs in distribution.

If the calculated F-statistic is large enough, based on the critical value from the F-distribution table (depending on the significance level), we reject the null hypothesis and conclude that there is a significant difference between the group means.

Relation to Hypothesis Testing:
ANOVA and the F-test: The F-test is the foundation of ANOVA. In ANOVA, the F-test is used to test whether there is a significant difference in the means of three or more groups. The idea is that if the variation between the groups is much larger than the variation within the groups, the null hypothesis can be rejected, suggesting that at least one group mean is different.

In One-Way ANOVA, for example, the F-test compares the variance of group means to the variance within each group.

In Regression Analysis, the F-test assesses the overall significance of the model (whether the model explains a significant portion of the variance in the dependent variable).

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to calculate and visualize Type 1 and Type 2 errors
def type_1_2_errors(population_mean, sample_size, population_std_dev, significance_level, true_mean):
    # Simulating a sample data from a normal distribution with the population mean
    sample_data_null = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)
    sample_mean_null = np.mean(sample_data_null)

    # Z-score calculation for the null hypothesis (H0: sample mean = population mean)
    z_score_null = (sample_mean_null - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Simulating a sample data from a normal distribution with the true mean (alternative hypothesis)
    sample_data_alt = np.random.normal(loc=true_mean, scale=population_std_dev, size=sample_size)
    sample_mean_alt = np.mean(sample_data_alt)

    # Z-score calculation for the alternative hypothesis (H1: sample mean ≠ true mean)
    z_score_alt = (sample_mean_alt - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Critical Z-value for the two-tailed test
    z_critical = stats.norm.ppf(1 - significance_level / 2)

    # Type 1 Error (α) - False Positive: Rejecting H0 when it is true
    type_1_error = stats.norm.cdf(-z_critical) + (1 - stats.norm.cdf(z_critical))

    # Type 2 Error (β) - False Negative: Failing to reject H0 when it is false (H1 is true)
    power = 1 - stats.norm.cdf(z_critical - (true_mean - population_mean) / (population_std_dev / np.sqrt(sample_size)))
    type_2_error = 1 - power

    # Print errors
    print(f"Sample mean under H0: {sample_mean_null}")
    print(f"Sample mean under H1: {sample_mean_alt}")
    print(f"Z-score under H0: {z_score_null}")
    print(f"Z-score under H1: {z_score_alt}")
    print(f"Type 1 Error (α): {type_1_error}")
    print(f"Type 2 Error (β): {type_2_error}")

    # Visualization
    plot_type_errors(z_score_null, z_score_alt, z_critical, type_1_error, type_2_error)

# Function to plot the Type 1 and Type 2 error regions
def plot_type_errors(z_score_null, z_score_alt, z_critical, type_1_error, type_2_error):
    # Plotting the normal distribution curve
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label="Standard Normal Distribution")

    # Shading the rejection regions for Type 1 Error (α)
    plt.fill_between(x, 0, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region (Type 1 Error)')

    # Shading the Type 2 Error region (Failing to Reject H0 when H1 is true)
    x_alt = np.linspace(z_critical - 1, 4, 1000)
    plt.fill_between(x_alt, 0, stats.norm.pdf(x_alt), color='green', alpha=0.3, label='Type 2 Error Region')

    # Mark the observed Z-scores for H0 and H1
    plt.axvline(z_score_null, color='blue', linestyle='dashed', label=f'Z-score under H0 = {z_score_null:.2f}')
    plt.axvline(z_score_alt, color='orange', linestyle='dashed', label=f'Z-score under H1 = {z_score_alt:.2f}')

    # Labels and title
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend(loc='upper right')

    # Display the plot
    plt.show()

# Define parameters for the test
population_mean = 100  # Population mean
population_std_dev = 15  # Population standard deviation
sample_size = 50  # Sample size
significance_level = 0.05  # Significance level
true_mean = 105  # True mean (Alternative hypothesis H1)

# Calculate and visualize Type 1 and Type 2 errors
type_1_2_errors(population_mean, sample_size, population_std_dev, significance_level, true_mean)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

# Function to calculate confidence interval for a sample mean
def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample std deviation
    sample_size = len(sample_data)

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the Z-score for the desired confidence level
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = z_score * standard_error

    # Confidence interval (lower bound and upper bound)
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage with simulated data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=100)  # Simulated sample data

# Calculate 95% confidence interval
confidence_level = 0.95
lower, upper = calculate_confidence_interval(sample_data, confidence_level)

# Display the result
print(f"The {confidence_level*100}% confidence interval is: ({lower:.2f}, {upper:.2f})")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior, likelihood, evidence):
    """
    Perform Bayesian inference using Bayes' Theorem.

    Parameters:
    prior (float): The prior probability P(H), the initial belief about the hypothesis.
    likelihood (float): The likelihood P(E|H), the probability of evidence given the hypothesis.
    evidence (float): The total probability P(E), the probability of observing the evidence.

    Returns:
    float: The posterior probability P(H|E), the updated belief about the hypothesis after observing evidence.
    """
    # Applying Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Example: Medical Test for Disease

# Prior: Probability of having the disease (1% of population has the disease)
prior = 0.01

# Likelihood: Probability of a positive test result given that the person has the disease (90% sensitivity)
likelihood = 0.90

# Evidence: Total probability of a positive test result (True Positives + False Positives)
# P(E) = P(E|H) * P(H) + P(E|¬H) * P(¬H)
# P(E|¬H): Probability of a positive test result given that the person does not have the disease (10% false positive rate)
false_positive_rate = 0.10
evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

# Calculate the posterior probability
posterior = bayesian_inference(prior, likelihood, evidence)

# Display the result
print(f"The posterior probability of having the disease given a positive test result is: {posterior:.4f}")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy import stats

# Function to perform F-test
def f_test(sample1, sample2):
    # Calculate variances of both samples
    var1 = np.var(sample1, ddof=1)  # Sample variance (ddof=1 for unbiased estimator)
    var2 = np.var(sample2, ddof=1)

    # Calculate the F-statistic (larger variance / smaller variance)
    if var1 > var2:
        F_statistic = var1 / var2
        df1 = len(sample1) - 1  # degrees of freedom for sample1
        df2 = len(sample2) - 1  # degrees of freedom for sample2
    else:
        F_statistic = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1

    # Calculate the p-value from the F-distribution
    p_value = 1 - stats.f.cdf(F_statistic, df1, df2)

    return F_statistic, p_value

# Simulating two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1 with mean=50, std=10
sample2 = np.random.normal(loc=50, scale=15, size=100)  # Sample 2 with mean=50, std=15

# Perform the F-test
F_statistic, p_value = f_test(sample1, sample2)

# Display the results
print(f"F-statistic: {F_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check the assumptions of ANOVA
def check_anova_assumptions(groups):
    # 1. Check Normality of each group using Shapiro-Wilk Test
    print("Normality Test (Shapiro-Wilk Test):")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: Statistic={stat:.4f}, P-value={p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i}: The data is not normally distributed (reject H0).\n")
        else:
            print(f"Group {i}: The data is normally distributed (fail to reject H0).\n")

    # 2. Check Equal Variance using Levene's Test
    print("Levene's Test for Equal Variance:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test Statistic={stat:.4f}, P-value={p_value:.4f}")
    if p_value < 0.05:
        print("The variances are significantly different (reject H0).\n")
    else:
        print("The variances are equal (fail to reject H0).\n")

    # 3. Visualize the data with histograms and box plots
    plt.figure(figsize=(12, 6))

    # Plot Histograms for Normality Check
    plt.subplot(1, 2, 1)
    for i, group in enumerate(groups, start=1):
        sns.histplot(group, kde=True, label=f'Group {i}')
    plt.title("Histograms with KDE for Normality Check")
    plt.legend()

    # Plot Boxplot for Equal Variance Check
    plt.subplot(1, 2, 2)
    sns.boxplot(data=groups)
    plt.title("Boxplot for Equal Variance Check")
    plt.xticks([0, 1, 2], [f'Group {i}' for i in range(1, len(groups)+1)])

    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# List of groups to check
groups = [group1, group2, group3]

# Check the assumptions for ANOVA
check_anova_assumptions(groups)
```

20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Simulating data for two-way ANOVA
np.random.seed(42)

# Factor A: Treatment (A1, A2)
# Factor B: Gender (Male, Female)
# Dependent Variable: Test Scores
factor_a = np.array(["A1"] * 50 + ["A2"] * 50)
factor_b = np.array(["Male"] * 25 + ["Female"] * 25 + ["Male"] * 25 + ["Female"] * 25)
scores = np.concatenate([
    np.random.normal(75, 10, 50),  # A1, Male
    np.random.normal(80, 10, 50),  # A1, Female
    np.random.normal(85, 10, 50),  # A2, Male
    np.random.normal(90, 10, 50)   # A2, Female
])

# Creating a DataFrame
df= pd.DataFrame({
    "Factor_A": factor_a,
    "Factor_B": factor_b,
    "Scores": scores})

# Performing Two-way ANOVA using statsmodels
model = ols('Scores ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = anova_lm(model, typ=2)
print(anova_table)

# Visualizing the interaction between factors using a line plot
plt.figure(figsize=(10, 6))
sns.pointplot(data=df, x="Factor_A", y="Scores", hue="Factor_B", markers=["o", "s"], ci=None)
plt.title('Interaction Plot for Two-way ANOVA')
plt.xlabel('Factor A (Treatment)')
plt.ylabel('Scores')
plt.legend(title="Factor B (Gender)")
plt.show()
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # n = 30, mean = 50, std = 10

# Step 2: Define the hypotheses
sigma_0_squared = 100  # Hypothesized population variance (e.g., 100)
n = len(sample_data)
sample_variance = np.var(sample_data, ddof=1)  # Sample variance

# Step 3: Calculate the test statistic
chi_square_stat = (n - 1) * sample_variance / sigma_0_squared

# Step 4: Determine the degrees of freedom (df) and significance level
df = n - 1
alpha = 0.05  # Significance level

# Find the critical value for Chi-square (for two-tailed test)
chi_square_critical_lower = stats.chi2.ppf(alpha / 2, df)
chi_square_critical_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Step 5: Compute the p-value
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Step 6: Print the results
print(f"Sample Variance: {sample_variance}")
print(f"Chi-square Statistic: {chi_square_stat}")
print(f"Degrees of Freedom (df): {df}")
print(f"Critical Values: Lower = {chi_square_critical_lower}, Upper = {chi_square_critical_upper}")
print(f"P-value: {p_value}")

# Step 7: Interpretation
if chi_square_stat < chi_square_critical_lower or chi_square_stat > chi_square_critical_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in population variance.")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

# Function to perform Z-test for proportions
def z_test_proportions(successes_1, n_1, successes_2, n_2, alpha=0.05):
    # Step 1: Calculate the sample proportions
    p1_hat = successes_1 / n_1
    p2_hat = successes_2 / n_2

    # Step 2: Calculate the combined proportion (P)
    p_combined = (successes_1 + successes_2) / (n_1 + n_2)

    # Step 3: Calculate the standard error of the difference in proportions
    standard_error = np.sqrt(p_combined * (1 - p_combined) * (1 / n_1 + 1 / n_2))

    # Step 4: Calculate the Z-statistic
    z_stat = (p1_hat - p2_hat) / standard_error

    # Step 5: Find the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Step 6: Print the results
    print(f"Sample 1 Proportion (p1_hat): {p1_hat}")
    print(f"Sample 2 Proportion (p2_hat): {p2_hat}")
    print(f"Combined Proportion (p_combined): {p_combined}")
    print(f"Z-statistic: {z_stat}")
    print(f"P-value: {p_value}")

    # Step 7: Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The proportions are not significantly different.")

# Example Data
successes_1 = 80  # Number of successes in group 1
n_1 = 200         # Sample size of group 1
successes_2 = 60  # Number of successes in group 2
n_2 = 180         # Sample size of group 2

# Perform the Z-test for proportions
z_test_proportions(successes_1, n_1, successes_2, n_2)
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform F-test for comparing variances
def f_test_variances(data1, data2, alpha=0.05):
    # Step 1: Calculate the variances of both datasets
    var1 = np.var(data1, ddof=1)  # ddof=1 for sample variance
    var2 = np.var(data2, ddof=1)

    # Step 2: Calculate the F-statistic (larger variance / smaller variance)
    F_statistic = var1 / var2 if var1 > var2 else var2 / var1

    # Step 3: Calculate the degrees of freedom for both samples
    df1 = len(data1) - 1
    df2 = len(data2) - 1

    # Step 4: Compute the p-value using the F-distribution
    p_value = 1 - stats.f.cdf(F_statistic, df1, df2)

    # Step 5: Print the results
    print(f"Variance of Data1: {var1}")
    print(f"Variance of Data2: {var2}")
    print(f"F-statistic: {F_statistic}")
    print(f"P-value: {p_value}")

    # Step 6: Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

    # Step 7: Visualize the distributions and F-statistic
    plt.figure(figsize=(10, 6))

    # Plot histogram of data
    plt.subplot(121)
    plt.hist(data1, alpha=0.7, label='Data 1', bins=15, color='blue')
    plt.hist(data2, alpha=0.7, label='Data 2', bins=15, color='orange')
    plt.title('Histogram of Data 1 and Data 2')
    plt.legend()

    # Plot the F-distribution
    plt.subplot(122)
    x = np.linspace(0, 5, 500)
    plt.plot(x, stats.f.pdf(x, df1, df2), label='F-distribution', color='green')
    plt.axvline(x=F_statistic, color='red', linestyle='dashed', label=f'F-statistic = {F_statistic:.2f}')
    plt.title('F-distribution with F-statistic')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Example Data: Simulate two random datasets
np.random.seed(0)
data1 = np.random.normal(loc=10, scale=5, size=100)  # Dataset 1 with mean=10, std=5
data2 = np.random.normal(loc=10, scale=7, size=100)  # Dataset 2 with mean=10, std=7

# Perform the F-test for variances
f_test_variances(data1, data2)
```

27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data: Suppose we have a categorical variable with 6 possible outcomes
np.random.seed(0)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=1000)

# Calculate observed frequencies
observed_freq = [np.sum(observed_data == i) for i in range(1, 7)]

# Define expected frequencies for a uniform distribution
expected_freq = [len(observed_data) / 6] * 6  # Uniform distribution, expected freq = total / number of categories

# Perform Chi-square test for goodness of fit
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

# Print the results
print(f"Observed frequencies: {observed_freq}")
print(f"Expected frequencies: {expected_freq}")
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The observed data does not follow a uniform distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows a uniform distribution.")

# Visualize the observed and expected frequencies
categories = [1, 2, 3, 4, 5, 6]

plt.bar(categories, observed_freq, alpha=0.7, label='Observed', color='blue', width=0.4, align='center')
plt.bar(categories, expected_freq, alpha=0.7, label='Expected', color='red', width=0.4, align='edge')

plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies (Chi-square Goodness of Fit)')
plt.legend()
plt.show()
```
